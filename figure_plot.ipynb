```python
plt.figure(0)
x = np.arange(1e-3, 5, 0.01)

kl = x * np.log(x)
lecam = 0.25* (1-x)**2/(1+x)
chi2 = x**2 -1
tv = 0.5 * np.abs(x-1)
js = (x+1) * np.log(2/(x+1)) + x* np.log(x)


plt.plot(x, lecam, linewidth=2, label='LeCam-divergence', alpha=.8)
plt.plot(x, js, linewidth=2, label='JS-divergence', marker='+', alpha=.8, markevery=50)
plt.plot(x, tv, linewidth=2, label='Total Variance', alpha=.8, marker = 6, markevery=50)
plt.plot(x, kl, '--', linewidth=2, label='KL-divergence', marker=',', alpha=.8)
plt.plot(x, chi2, '-.', linewidth=1, label=r'$\chi^2$-divergence', alpha=.8)

plt.rc('text', usetex=False)
plt.xlabel(r'$\frac{P(x)}{Q(x)}$', fontsize=12)
plt.ylabel(r'$f(\frac{P(x)}{Q(x)})$', fontsize=12)

plt.ylim(-1, 5)
plt.gca().legend(loc=0, prop={'size': 11})
# plt.savefig(os.open('lecam.pdf', 'wb'), format='pdf', dpi=200, bbox_inches='tight')
```
